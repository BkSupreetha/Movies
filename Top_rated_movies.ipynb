```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
data = pd.read_csv('movies.csv')
data
```

```python
data.columns
```

```python
#setting first column as index
data = data.set_index('Unnamed: 0')
```

```python
data
```

```python
#renaming the index column
data = data.rename_axis('Index')
```

```python
#checking for null values
data.isnull().sum()
```

```python
#selecting non adult movies
print((data['adult']=='False'), data[['title', 'id']])
(data['adult']=='False').count()
```

```python
#selecting non adult movies using lambda function
Non_Adult_movie =  data[['title', 'id']].apply(lambda x: x[data['adult']=='False'])
Non_Adult_movie
```

```python
Adult_movie =  data[['title', 'id']].apply(lambda x: x[data['adult']!='False'])
Adult_movie
```

```python
print('Total no. of titles ',data['title'].count())
unique_titles = data['title'].unique()
print('Total no. of unique titles: ',unique_titles.size)
```

# Since we can see that there is difference between no. of titless and no. of unique titles it is obvious that few movie titles have been repeated,  lets explore those itles and their count

```python
#checking for title values and their count of appearance
title_recurring_counts = data['title'].value_counts()
title_recurring_counts
```

```python
#printing repeated title names
repeated_titles = title_recurring_counts[title_recurring_counts > 1]
repeated_titles
```

```python
#Deleting repeated titles
data.drop_duplicates(subset='title', keep='first', inplace=True)
data
```
